Задание 1<br>
Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.<br>
Напишите код, который проверяет какая из этих строк длиннее.<br>

Примеры работы программы:<br>1.<br>

phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики’<br>
phrase_2 = ‘640Кб должно хватить для любых задач. Билл Гейтс (по легенде)’<br>

Результат:<br>
Фраза 1 длиннее фразы 2<br> 2.<br>

phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'<br>
phrase_2 = ‘Насколько проще было бы писать программы, если бы не заказчики’<br>

Результат:<br>
Фраза 2 длиннее фразы 1<br> 3.<br>

phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики’<br>
phrase_2 = ‘Насколько проще было бы писать программы, если бы не заказчики’<br>

Результат:<br>
Фразы равной длины<br>

```python
from calendar import month

phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

phrase_1_len = len(phrase_1)
phrase_2_len = len(phrase_2)

if phrase_1_len > phrase_2_len:
    print('Фраза 1 длиннее фразы 2')
elif phrase_1_len == phrase_2_len:
    print('Фразы равной длины')
else:
    print('Фраза 2 длиннее фразы 1')
```

Задание 2<br>
Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или<br> обычным.<br>

Пример работы программы: 1.<br>

year = 2020<br>
Результат:<br>
Високосный год<br>

year = 2019<br>
Результат:<br>
Обычный год<br>

```python
year = 2020
if(year % 4 == 0 and year % 100 != 0 or year % 400 == 0):
    print("високосный год")
else:
    print("обычный год")
```

Задание 3<br>
Необходимо написать программу, которая будет запрашивать у пользователя месяц и дату рождения и выводить соответствующий знак зодиака.<br>

Пример работы программы: <br>1.<br>

30<br>

Введите месяц:<br>

Результат:<br>
Ваш знак зодиака: Дева<br>

Введите день:<br>
29<br>

Введите месяц:<br>
Октябрь<br>
Результат:<br>
Ваш знак зодиака: Скорпион<br>

```python
day = int(input('Введите день:'))
month = str(input('Введите месяц:'))

print('Результат:')
res = 'Ваш знак зодиака: '

month_dict = {
        'январь': 1, 'февраль': 2, 'март': 3, 'апрель': 4,
        'май': 5, 'июнь': 6, 'июль': 7, 'август': 8,
        'сентябрь': 9, 'октябрь': 10, 'ноябрь': 11, 'декабрь': 12
    }

month = month_dict.get(month.lower())

if (month == 12 and day >= 22) or (month == 1 and day <= 19):
        print(res + 'Козерог')
elif (month == 1 and day >= 20) or (month == 2 and day <= 18):
        print(res + 'Водолей')
elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        print(res + 'Рыбы')
elif (month == 3 and day >= 21) or (month == 4 and day <= 19):
        print(res + 'Овен')
elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        print(res + 'Телец')
elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        print(res + 'Близнецы')
elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        print(res + 'Рак')
elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        print(res + 'Лев')
elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        print(res + 'Дева')
elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        print(res + 'Весы')
elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        print(res + 'Скорпион')
elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        print(res + 'Стрелец')
else:
        print('Ошибка: некорректная дата')
```

Задание 4<br>
Вам нужно написать программу для подбора упаковок по размерам товара. Размеры (ширина, длина, высота) хранятся в переменных (в сантиметрах):<br>

Используйте следующие правила:<br>

если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на экран “Коробка №1”;<br>
если хотя бы одно из измерений больше 2 метров, то выводите “Упаковка для лыж”;<br>
если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите “Коробка №2”;<br>
во всех остальных случаях выводите “Коробка №3”.<br>
Пример работы программы:<br>
<br>1.<br>

width = 15<br>
length = 55<br>
height = 15<br>
Результат:<br>
Коробка №3<br>

width = 45<br>
length = 205<br>
height = 45<br>
Результат:<br>
Упаковка для лыж<br>

```python
width = 15
length = 55
height = 15
if width > 200 or length > 200 or height > 200:
    print('Упаковка для лыж')
elif width <= 15 and length <= 15 and height <= 15:
    print('Коробка №1')
elif 15 < width < 50 or 15 < length < 50 or 15 < height < 50:
    print('Коробка №2')
else:
    print('Коробка №3')
```
