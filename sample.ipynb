Задание 1<br>
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

среднюю букву, если число букв в слове нечетное;<br>
две средних буквы, если число букв четное.<br>
Примеры работы программы:<br>

word = 'test’<br>
Результат:<br>
es<br>

word = 'testing’<br>
Результат:<br>
t

```python
word = 'test'
length = len(word)
mid = length // 2
if (length % 2  > 0):
    print(word[mid])
else:
    print(word[mid - 1] + word[mid])
```

Задание 2<br>
Напишите программу, которая последовательно запрашивает у пользователя числа<br> (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы: 1.<br>

Введите число: 1<br>

Введите число: 4<br>

Введите число: 6<br>

Введите число: 0<br>
Результат:11<br>

Введите число: 0<br>
Результат:0

```python
summ = 0
loop = True
while loop:
    number = int(input('введите цифру: '))
    if number == 0:
        loop = False
        print(summ)
    else:
        summ = summ + number
```

Задание 3 <br>
Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек.<br>
Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами <br> после сортировки! Но мы не будем никого знакомить, если кто-то может остаться без пары:<br>

Примеры работы программы: 1.<br>

boys = [‘Peter’, ‘Alex’, ‘John’, ‘Arthur’, ‘Richard’]<br>
girls = [‘Kate’, ‘Liza’, ‘Kira’, ‘Emma’, ‘Trisha’]<br>
Результат:<br>

Идеальные пары:<br>
Alex и Emma<br>
Arthur и Kate<br>
John и Kira<br>
Peter и Liza<br>
Richard и Trisha<br>
boys = [‘Peter’, ‘Alex’, ‘John’, ‘Arthur’, ‘Richard’, ‘Michael’]<br>
girls = [‘Kate’, ‘Liza’, ‘Kira’, ‘Emma’, ‘Trisha’]<br>
Результат:<br>
Внимание, кто-то может остаться без пары!<br>

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys) == len(girls):
    sorted_boys = sorted(boys)
    sorted_girls = sorted(girls)
    for i in range(len(sorted_boys)):
        print(sorted_boys[i] + ' и ' + sorted_girls[i])
else:
    print('Внимание, кто-то может остаться без пары!')
```

Задание 4<br>
У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах <br>за произвольный период по странам (структура данных в примере). Необходимо написать код, который рассчитает среднюю температуру за период<br> в Цельсиях(!) для каждой страны.

Пример работы программы:<br>

countries_temperature = [<br>
[‘Таиланд’, [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],<br>
[‘Германия’, [57.2, 55.4, 59, 59, 53.6]],<br>
[‘Россия’, [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],<br>
[‘Польша’, [50, 50, 53.6, 57.2, 55.4, 55.4]]<br>
]<br>
Результат:<br>

Средняя температура в странах:<br>
Таиланд - 23.9 С<br>
Германия - 13.8 С<br>
Россия - 3.7 С<br>
Польша - 12.0 С<br>

```python
countries_temperature = [
['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['Германия', [57.2, 55.4, 59, 59, 53.6]],
['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

print('Средняя температура в странах:')

for i in range(len(countries_temperature)):
    country_name = countries_temperature[i][0]
    temperature_arr = countries_temperature[i][1]

    mid_temp = round((sum(temperature_arr) / len(temperature_arr)) - 32 / 1.8, 1)

    print(country_name + ' - ' + str(mid_temp) + 'C')
```
