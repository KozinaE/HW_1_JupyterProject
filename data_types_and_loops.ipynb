Задание 1<br>
Дана переменная, в которой хранится словарь, содержащий гео-метки для каждого пользователя<br> (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран множество<br> уникальных гео-меток всех пользователей.

Пример работы программы:<br>

<code/>ids = {‘user1’: [213, 213, 213, 15, 213],
‘user2’: [54, 54, 119, 119, 119],
‘user3’: [213, 98, 98, 35]}</code> <br>
Результат: {98, 35, 15, 213, 54, 119}

```python
ids = {'user1': [213, 213, 213, 15, 213],
'user2': [54, 54, 119, 119, 119],
'user3': [213, 98, 98, 35]}

unic_geo_tags = set()

for geo_tag in ids.values():
    unic_geo_tags.update(geo_tag)
print('Результат: ' + str(unic_geo_tags))
```

Задание 2<br>
Дана переменная, в которой хранится список поисковых запросов пользователя<br> (пример структуры данных приведен ниже, но запросы потенциально могут быть любые). Вам необходимо<br> написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

Пример работы программы:<br>
<code/>
queries = [
‘смотреть сериалы онлайн’,
‘новости спорта’,
‘афиша кино’,
‘курс доллара’,
‘сериалы этим летом’,
‘курс по питону’,
‘сериалы про спорт’,
]</code><br>
Результат:<br>

Поисковых запросов, содержащих 2 слов(а): 42.86%<br>
Поисковых запросов, содержащих 3 слов(а): 57.14%

```python
queries = [
'смотреть сериалы онлайн',
'новости спорта',
'афиша кино',
'курс доллара',
'сериалы этим летом',
'курс по питону',
'сериалы про спорт'
]
len_count_map = {}

for query in queries:
    query_len = len(query.split(' '))
    if query_len not in len_count_map:
        len_count_map[query_len] = 1
    else:
        len_count_map[query_len] += 1

total = len(queries)
for word_count in sorted(len_count_map.keys()):
    count = len_count_map[word_count]
    percentage = (count / total) * 100
    print(f'Поисковых запросов, содержащих {word_count} слов(а): {percentage:.2f}%')
```

Задание 3<br>
Дана переменная, в которой хранится информация о затратах и доходе рекламных кампаний по различным источникам. Необходимо дополнить исходную структуру показателем ROI,<br> который рассчитаем по формуле: ((revenue / cost) - 1) * 100<br>

Пример работы программы:<br>

results = {<br>
‘vk’: {‘revenue’: 103, ‘cost’: 98},<br>
‘yandex’: {‘revenue’: 179, ‘cost’: 153},<br>
‘facebook’: {‘revenue’: 103, ‘cost’: 110},<br>
‘adwords’: {‘revenue’: 35, ‘cost’: 34},<br>
‘twitter’: {‘revenue’: 11, ‘cost’: 24},<br>
}<br>
Результат:<br>

{‘adwords’: {‘revenue’: 35, ‘cost’: 34, ‘ROI’: 2.94},<br>
‘facebook’: {‘revenue’: 103, ‘cost’: 110, ‘ROI’: -6.36},<br>
‘twitter’: {‘revenue’: 11, ‘cost’: 24, ‘ROI’: -54.17},<br>
‘vk’: {‘revenue’: 103, ‘cost’: 98, ‘ROI’: 5.1},<br>
‘yandex’: {‘revenue’: 179, ‘cost’: 153, ‘ROI’: 16.99}}<br>

```python
results = {
'vk': {'revenue': 103, 'cost': 98},
'yandex': {'revenue': 179, 'cost': 153},
'facebook': {'revenue': 103, 'cost': 110},
'adwords': {'revenue': 35, 'cost': 34},
'twitter': {'revenue': 11, 'cost': 24},
}

for key in results.keys():
    data = results[key]
    roi = round(((data['revenue'] / data['cost']) - 1) * 100, 2)
    data['ROI'] = roi

sorted_results = dict(sorted(results.items()))

for source, data in sorted_results.items():
    print(f"'{source}': {data}")
```

Задание 4<br>
Дана переменная, в которой хранится статистика рекламных каналов по объемам продаж<br> (пример структуры данных приведен ниже). Напишите программу, которая возвращает название канала с максимальным объемом продаж.<br>

Пример работы программы:<br>

<code/>stats = {‘facebook’: 55, ‘yandex’: 115, ‘vk’: 120, ‘google’: 99, ‘email’: 42, ‘ok’: 98}</code><br>

Результат: Максимальный объем продаж на рекламном канале: vk

```python
stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}

max_channel = max(stats, key = stats.get)
print(f'Максимальный объем продаж на рекламном канале: {max_channel}')
```
